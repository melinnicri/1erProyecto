Abrir Archivo comprimido de datos de items de usuario:

```python
# Importar librerías
import gzip
import pandas as pd
import json
import numpy as np
```

```python
# Descomprimir el archivo gzip
with gzip.open('DataBase/users_items.json.gz', 'rb') as f:
    data = f.readlines()
```

```python
# Convertir las líneas a registros JSON   1m 33 seg
records = [eval(line.strip()) for line in data]
```

```python
# Crear el DataFrame a partir de los registros
df_items= pd.DataFrame(records)
```

```python
# Revisar data 88310 rows × 5 columns
df_items
```

```python
#Desanidar la columna anidada "Items", 12,3 seg
data_desanidada = []

for index, row in df_items.iterrows():
    user_id = row['user_id']
    items_count = row['items_count']
    steam_id = row['steam_id']
    user_url = row['user_url']
    items = row['items']
    
    for i in items:   
        new_row = {
        'user_id': user_id,
        'items_count': items_count,
        'steam_id' : steam_id,
        'user_url' : user_url,
        'item_id': i.get('item_id', ''),
        'item_name': i.get('item_name', ''),
        'playtime_forever': i.get('playtime_forever', ''),
        'playtime_2weeks': i.get('playtime_2weeks', '')
        }
        
        data_desanidada.append(new_row)
```

```python
#DataFrame desanidada     8,5 seg
df_i_desanidado = pd.DataFrame(data_desanidada)
```

```python
df_i_desanidado    # 5153209 rows × 8 columns
```

```python
#Cuántos valores NaN hay por columna
df_i_desanidado.isna().sum()
```

```python
#Crear una nueva dataframe con user_id, items_count, steam_id, item_id, item_name, playtime_forever, sin duplicados
df_usuario = df_i_desanidado[["user_id", "items_count", "steam_id", "item_id", "item_name", "playtime_forever"]]
```

```python
#Guardamos en un archivo .csv
df_usuario.to_csv('Third_Users.csv', index=False)
```
