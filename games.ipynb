Abrir Archivo comprimido de datos de juego:

```python
# Importar librerías
import gzip
import pandas as pd
```

```python
# Descomprimir el archivo gzip
with gzip.open('DataBase/steam_games.json.gz', 'rb') as f:
    file_content = f.read()
```

```python
# Leer el archivo json con pandas
data = pd.read_json(file_content.decode("utf-8"), lines=True)
```

```python
#Borramos las celdas con NaNs (más de tres, 88 mil NaNs celdas completas)
data = data.dropna(thresh=3)
```

```python
data #32135 rows × 13 columnas
```

```python
# Cuántas columnas con nulos, hasta con tres por fila
data.isnull().sum()  
```

```python
# Revisar filas con nulos
print("\nfilas completamente nulas: ", data.isna().all(axis=1).sum()) 
```

```python
#Extraer año
data["release_year"] = data["release_date"].str.extract(r"(\d{4})")
```

```python
data.head(5)
```

```python
# Nombre de columnas
data.columns
```

```python
df = pd.DataFrame(data)
```

```python
#Crear una nueva dataframe con "genres", "release_year", "id", "app_name", "title", "tags", "developer", sin duplicados
df_games = df[["genres", "release_year", "id", "app_name", "title", "tags", "developer"]]
```

```python
df_games
```

```python
print(df_games.dtypes)
```

```python
#Convertir la columna genres a string
df_games["genres"] = df["genres"].astype(str)
```

```python
#Dividir genres con app_name archivo aparte
df_games = df.explode("genres").reset_index(drop=True)
```

```python
#Crear una nueva dataframe con genres, release_year, id, app_name, sin duplicados
df_genres = df_games[["genres", "release_year", "id", "app_name"]]
```

```python
#Guardamos nueva data en un archivo .csv
df_genres.to_csv('First_Game_genres.csv', index=False)
```

```python
#Crear una nueva dataframe con title tags developer id, sin duplicados HAY QUE DESANIDAR!!!
df_tags = df_games[["title", "tags", "developer", "id"]]
```

```python
#Guardamos nueva data en un archivo .csv
df_tags.to_csv('First_Game_tags.csv', index=False)
```
